```python
import pandas as pd
df = pd.read_csv('DataSet/course_recommendation.csv')
# convert column 'Rating' to numeric
df['Rating'] = pd.to_numeric(df['Rating'])
df
```

```python
label = {
    -2: 'Hated',
    -1: 'Hated',
    0: 'Liked',
    1: 'Liked',
    2: 'Liked'
}
```

```python
# Overall
val_cnt = df['Rating'].map(label).value_counts()
val_perc = df['Rating'].map(label).value_counts(normalize=True) * 100
overall = pd.concat([val_cnt, val_perc], axis=1, keys=['counts', 'percentage'])
overall
```

```python
# set of features in df
features = set(df.columns) - {'Rating'}
```

```python
score = {}
for feature in features:
    no = df[df[feature] == 'n']['Rating'].map(label).value_counts()
    no
    yes = df[df[feature] == 'y']['Rating'].map(label).value_counts()
    yes

    score[feature] = no.max() + yes.max()

# find feature with max score
f = max(score, key=score.get)

for feature in features:
    print(feature, score[feature])
```
